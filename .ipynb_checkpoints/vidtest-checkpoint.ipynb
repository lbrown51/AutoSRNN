```python
import imageio
import numpy as np
from PIL import Image
import glob
```

```python
fns = glob.glob('./walk/*')
```

```python
vid = imageio.read(fns, 'FFMPEG')
```

```python
vids = np.array(list(imageio.read(fns[0], 'FFMPEG').iter_data()))
```

```python
for fn in fns[1:]:
    vid = np.array(list(imageio.read(fn, 'FFMPEG').iter_data()))
    try:
        vids = np.concatenate((vids, vid), axis=0)
    except ValueError:
        continue
```

```python
np.save('./walk_np', vids)
```
